```python
import torch
from torchvision import models, transforms
import numpy as np
from lime import lime_image
import matplotlib.pyplot as plt
from PIL import Image
import cv2
```

```python
IMG_SIZE = (224, 224)
```

```python
# Load the pre-trained ResNet model
model = models.resnet50(pretrained=True)
model.eval()

def predict_fn(images):
    images_tensor = torch.stack([preprocess_image_from_array(img) for img in images])
    with torch.no_grad():
        outputs = model(images_tensor)
        probs = torch.nn.functional.softmax(outputs, dim=1)
    return probs.numpy()


def preprocess_image_from_array(image_array):
    transform = transforms.Compose([
        transforms.ToPILImage(),
        transforms.Resize(IMG_SIZE),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])  #ResNet normalization
    ])
    image = transform(image_array)
    return image
```

```python
#hotpot
base_path = '/Users/juliajersova/Library/CloudStorage/OneDrive-AZCollaboration/study/2 semestr/Diplom/data/raw_data1/hotpot/'
img_path = base_path + 'n07590611_3587.JPEG'
img = Image.open(img_path).convert("RGB")
img_path_bounding_box = "/Users/juliajersova/Library/CloudStorage/OneDrive-AZCollaboration/study/2 semestr/Diplom/outputs/bounding_boxes_OPENAI/hotpot_vase/n07590611_3587_with_box.jpg"
#n07590611_3587.JPEG,Hot pot with various ingredients and condiments.
```

```python
explainer = lime_image.LimeImageExplainer()

explanation = explainer.explain_instance(
    np.array(img),
    predict_fn,
    top_labels=1,
    hide_color=0,
    num_samples=1000
)
```

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 6))
img_rgb = Image.open(img_path_bounding_box).convert("RGB")

ax[0].imshow(img_rgb)
ax[0].axis('off')
ax[0].set_title('Bounding box', fontsize=24)


ax[1].imshow(explanation.segments, cmap="viridis")
ax[1].axis('off')
ax[1].set_title('LIME Heatmap', fontsize=24)

plt.tight_layout()
plt.show()
```
